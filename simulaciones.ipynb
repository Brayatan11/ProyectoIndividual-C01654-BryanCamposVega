```python
def check_misplaced_letters(word: str) -> bool:
        
        word = re.sub(r'[^A-Z]', '', word.upper())
        assert len(word) == 5, 'Word must be 5 characters long'
        
        # Break into letters
        list_word = list(word)
        
        # Get indices (1 indexed) of positions that have not yet been solved
        not_solved = [key for key, value in dict_letters.items() if value is None]
        
        # Filter list of words for those not yet solved, removing 1 from the index as our letters are 1 indexed
        list_word_unsolved = [list_word[i-1] for i in not_solved]
        
        # Check counts
        dict_count_letters = Counter(list_word_unsolved)
        
        # Compare to dictionary of misplaced letters
        valid = True
        for check_key, check_value in dict_misplaced_letters.items():
            if dict_count_letters[check_key] < check_value:
                valid = False
                
        return valid

def calcular_probabilidad_palabra(palabra, probabilidades_todas_posiciones):
    probabilidad_palabra = 1.0
    for i, letra in enumerate(palabra):
        posicion = i + 1
        probabilidad_letra = probabilidades_todas_posiciones.get(posicion, {}).get(letra, 0)
        probabilidad_palabra += probabilidad_letra
    return probabilidad_palabra

def calcular_probabilidades_palabras(palabras, probabilidades_todas_posiciones):
    lista_probabilidades = []
    for palabra in palabras:
        probabilidad = calcular_probabilidad_palabra(palabra, probabilidades_todas_posiciones)
        lista_probabilidades.append((palabra, probabilidad))
    lista_probabilidades.sort(key=lambda x: x[1], reverse=True)
    return lista_probabilidades

def conteo_letras(lista):
    arr_words_5l = np.array([list(w) for w in lista])
    df_words_5l = pd.DataFrame(data=arr_words_5l, columns=[f'letter_{i+1}' for i in range(5)])
    df_words_5l['word'] = lista
    
    test_dict_letter_counts = Counter()
    for i in range(5):
        test_dict_letter_counts[i+1] = Counter(df_words_5l[f'letter_{i+1}'])

    total_palabras = len(df_words_5l)
    probabilidades_todas_posiciones = {}
    for i in range(5):
        probabilidades_todas_posiciones[i+1] = {letra: conteo / total_palabras for letra, conteo in test_dict_letter_counts[i+1].items()}

    return probabilidades_todas_posiciones

def generar_estados(palabra_escogida, palabra_objetivo):
    estados = ['?' for _ in range(5)]
    palabra_objetivo_contada = Counter(palabra_objetivo)
    
    # Primero, marcar 'V' para letras correctas en la posición correcta
    for i in range(5):
        if palabra_escogida[i] == palabra_objetivo[i]:
            estados[i] = 'V'
            palabra_objetivo_contada[palabra_escogida[i]] -= 1
    
    # Luego, marcar 'Y' para letras correctas en la posición incorrecta
    for i in range(5):
        if estados[i] == '?':  # Solo mirar las letras que no han sido marcadas todavía
            if palabra_escogida[i] in palabra_objetivo_contada and palabra_objetivo_contada[palabra_escogida[i]] > 0:
                estados[i] = 'Y'
                palabra_objetivo_contada[palabra_escogida[i]] -= 1
    
    return ''.join(estados)

def simular_juego(list_possible_answers):

    max_intentos = 6
    intentos = 0
    
    probabilidades_todas_posiciones = conteo_letras(list_possible_answers)
    df_filtradas = df_words_5l
    palabras_filtradas = df_words_5l['word']
    
    dict_letters = defaultdict(str)
    for i in range(5):
        dict_letters[i+1] = None
    
    possible_letters = list(string.ascii_uppercase)
    resultados = {}
            
    for palabra_objetivo in list_possible_answers:
        
        df_filtradas = df_words_5l
        probabilidades_todas_posiciones = conteo_letras(list_possible_answers)
        palabras_filtradas = df_words_5l['word']
        lista_probabilidades = calcular_probabilidades_palabras(palabras_filtradas, probabilidades_todas_posiciones)
        possible_letters = list(string.ascii_uppercase)
        intentos = 0
        palabra_escogida = lista_probabilidades[0][0]

        dict_letters = defaultdict(str)
        for i in range(5):
            dict_letters[i+1] = None
            
        while intentos < max_intentos:
            estados = generar_estados(palabra_escogida, palabra_objetivo)
            
            if estados == 'VVVVV':
                resultados[palabra_objetivo] = intentos + 1
                intentos = 8
                break

            if intentos == 5 and estados != 'VVVVV':
                resultados[palabra_objetivo] = 7
                intentos = 6
                break
            
            # Convert palabra y estados into list of letters
            lista_palabra = list(palabra_escogida.upper())
            lista_estados = list(estados.upper())

            # Zip with results
            df_guess_results = pd.DataFrame(data=list(zip(lista_palabra, lista_estados)),
                                            columns=['letra', 'estado'],
                                            index=np.arange(1,6))
            
            dict_letters = defaultdict(str)
            for i in range(5):
                dict_letters[i+1] = None
        
            dict_misplaced_letters = Counter()
        
            already_solved = [key for key, value in dict_letters.items() if value is not None]

            # SOLVED
            df_letras_ver = df_guess_results.query('estado == "V"')
            if df_letras_ver.shape[0] > 0:
                for idx, row in df_letras_ver.iterrows():
                    
                    # Prevent updates for previously solved letters
                    if idx in already_solved:
                        pass
                    else:
                        corr_letter = row['letra']
                        dict_letters[idx] = corr_letter
                    
                        # If correct letter was previously guessed as a misplaced letter, remove it
                        if corr_letter in dict_misplaced_letters.keys():
                            dict_misplaced_letters[corr_letter] -= 1
                            
                        # And filter dataframe of possible words
                        df_filtradas = df_filtradas.query(f'letter_{idx}=="{corr_letter}"')

        
            # Add misplaced letters to our list, if it's a new letter
            df_letras_amar = df_guess_results.query('estado == "Y"')
            if df_letras_amar.shape[0] > 0:
                
                # Filter dataframe to remove any words that have the misplaced letter in that column
                for idx, row in df_letras_amar.iterrows():
                    mispl_letter = row['letra']
                    df_filtradas = df_filtradas.query(f'letter_{idx}!="{mispl_letter}"')  
                
                # Check how many we have of each letter that's misplaced
                guess_mispl_letters = df_letras_amar['letra'].values
                dict_guess_mispl_letters = Counter(guess_mispl_letters)
                
                dict_misplaced_letters = Counter(dict_misplaced_letters)
                # Then update our dictionary of misplaced letters
                for key, value in dict_guess_mispl_letters.items():
                    dict_misplaced_letters[key] = value   
                
                vect_check_misplaced_letters = np.vectorize(check_misplaced_letters)
                with pd.option_context('mode.chained_assignment', None):
                    df_filtradas.loc[:, 'valid'] = vect_check_misplaced_letters(df_filtradas['word'])
                df_filtradas = df_filtradas.query('valid == True')
                df_filtradas = df_filtradas.drop('valid', axis=1) 
        
            # Remove any incorrect letters from the list to guess from, if letter isn't in misplaced list
            df_wrong_answers = df_guess_results.query('estado == "?"')
            if df_wrong_answers.shape[0] > 0:

                for idx, row in df_wrong_answers.iterrows():
                    inc_letter = row['letra']
                    df_filtradas = df_filtradas.query(f'letter_{idx}!="{inc_letter}"')

                # Ensure we don't double count
                for l in df_wrong_answers['letra'].unique():
                    if dict_misplaced_letters[l] == 0  and l in possible_letters:
                        possible_letters.remove(l)
        
            # Finally, update list of possible 5 letter words by removing all rows where
            # for letters yet to be guessed, they don't fall in the list of possible letters
            yet_to_solve = [key for key, value in dict_letters.items() if value is None]
            #print(df_filtradas)
            for position in yet_to_solve:
                
                # Check all letters in a given position
                position_letters = df_filtradas[f'letter_{position}']
                
                # Return a boolean list of whether that list is in the possible values or not
                position_in_possible_letters = [l in possible_letters for l in position_letters]
                
                # Filter
                df_filtradas = df_filtradas[position_in_possible_letters].copy(deep=True)

            probabilidades_todas_posiciones = conteo_letras(df_filtradas['word'])
            lista_probabilidades_filtradas = calcular_probabilidades_palabras(df_filtradas['word'], probabilidades_todas_posiciones)
            palabra_escogida = lista_probabilidades_filtradas[0][0]
            intentos += 1

    
        print(f'{palabra_objetivo}: {resultados[palabra_objetivo]}')   
                
    
    return resultados


# Simular el juego para todas las palabras


# Imprimir los resultados
# for palabra, intentos in resultados_simulacion.items():
#     print(f"La palabra '{palabra}' fue adivinada en {intentos} intentos.")
```
